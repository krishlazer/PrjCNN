```python
import tensorflow as tf
import tensorflow.keras
import tensorflow.keras.models
from tensorflow.keras.layers import Dense, Conv2D, Flatten, MaxPool2D, Activation
import os
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.optimizers import Adam
import numpy as np
```

```python
train = ImageDataGenerator(rescale=1/255)
```

```python
validation = ImageDataGenerator(rescale=1/255)
```

```python
train_dataset = train.flow_from_directory('E:/KrishM/ML/IMG/gsearchcars/train', target_size=(100, 100), batch_size = 5, class_mode = 'binary')
```

```python
validation_dataset = validation.flow_from_directory('E:/KrishM/ML/IMG/gsearchcars/validation', target_size=(100, 100), batch_size = 5, class_mode='binary')
```

```python
train_dataset.class_indices
```

```python
validation_dataset.class_indices
```

```python
train_dataset.classes
```

```python
validation_dataset.classes
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3, 3), activation = 'relu', input_shape = (100, 100, 3)),
    tf.keras.layers.MaxPool2D(2, 2),
    
    tf.keras.layers.Conv2D(32, (3, 3), activation = 'relu'),
    tf.keras.layers.MaxPool2D(2, 2),
    
    tf.keras.layers.Conv2D(64, (3, 3), activation = 'relu'),
    tf.keras.layers.MaxPool2D(2, 2),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')  
    
])
```

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(train_dataset, epochs = 20, validation_data = validation_dataset)
```

```python
testcars = 'E:/KrishM/ML/IMG/gsearchcars/test'
for ci in os.listdir(testcars):
    img = image.load_img(testcars+'/' + ci, target_size = (100, 100))
    plt.imshow(img)
    plt.show()

    X = image.img_to_array(img)
    X = np.expand_dims(X, axis=0)
    images = np.vstack([X])
    carval = model.predict(images)
    if carval == 0:
        print('classic')
    else:
        print('sedan') 
    
```
